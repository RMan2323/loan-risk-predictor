```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import os
import matplotlib.pyplot as plt
```

```python
train_files = [
    "One-Hot/Train_Orig_OH.csv",
    "One-Hot/Scaled/Train_Scaled_All_OH.csv",
    "One-Hot/Scaled/Train_Scaled_Cont_OH.csv",
    "One-Hot/MinMax/Train_MM_OH.csv",
    "One-Hot/MinMax/train_OH_MM_PCA15.csv",
    "One-Hot/MinMax/train_OH_MM_PCA20.csv",
    "One-Hot/MinMax/train_OH_MM_PCA25.csv",
    "One-Hot/MinMax/train_OH_MM_PCA30.csv",
    "One-Hot/MinMax/train_OH_MM_PCA35.csv",
    "IntClasses/Train_Orig_Int.csv",
    "IntClasses/Scaled/Train_Scaled_All_Int.csv",
    "IntClasses/Scaled/Train_Scaled_Cont_Int.csv",
    "IntClasses/MinMax/Train_MM_Int.csv",
    "IntClasses/MinMax/train_Int_MM_PCA10.csv",
    "IntClasses/MinMax/train_Int_MM_PCA15.csv",
    "IntClasses/MinMax/train_Int_MM_PCA20.csv",
    "IntClasses/MinMax/train_Int_MM_PCA25.csv",
]
```

```python
from sklearn.metrics import mean_squared_error, r2_score

def evaluate_regression(y_true, y_pred, set_name=""):
    print(f"\nPerformance Metrics on {set_name} Data:")
    mse = mean_squared_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    print(f"Mean Squared Error (MSE): {mse:.4f}")
    print(f"R-squared (R²): {r2:.4f}")
    print("-" * 50)
    
    return mse, r2
```

```python
results = []

for path in train_files:
    print("\n" + "="*80)
    print(f"Processing dataset: {path}")
    
    train_df = pd.read_csv(path)
    valid_df = pd.read_csv(path.replace("Train", "Valid").replace("train", "valid"))
    test_df = pd.read_csv(path.replace("Train", "Test").replace("train", "test"))

    # Drop LoanApproved, keep RiskScore as target
    X_train = train_df.drop(columns=["LoanApproved", "RiskScore"])
    y_train = train_df["RiskScore"]

    X_valid = valid_df.drop(columns=["LoanApproved", "RiskScore"])
    y_valid = valid_df["RiskScore"]

    X_test = test_df.drop(columns=["LoanApproved", "RiskScore"])
    y_test = test_df["RiskScore"]

    # Train the model
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)

    name = os.path.basename(path)
    evaluate_regression(y_valid, model.predict(X_valid), "Validation")
    mse, r2 = evaluate_regression(y_test, model.predict(X_test), "Test")

    results.append({
        "Dataset": name,
        "MSE": mse,
        "R2": r2
    })

    # Optional: Plot top 10 features (only for non-PCA)
    if "PCA" not in path:
        importances = pd.Series(model.feature_importances_, index=X_train.columns)
        importances.nlargest(10).plot(kind='barh', title=f"Top 10 Features - {name}", figsize=(8, 5))
        plt.xlabel("Importance")
        plt.tight_layout()
        plt.show()

# Final summary
results_df = pd.DataFrame(results)
results_df
```

```python
#for checking if more trees would have helped or not
#They don't exactly help more and it starts plateauing afterwards so 100 does the job and efficient too
for n in [100, 200, 300, 500]:
    model = RandomForestRegressor(n_estimators=n, random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_valid)
    mse = mean_squared_error(y_valid, y_pred)
    r2 = r2_score(y_valid, y_pred)
    print(f"{n} trees -> MSE: {mse:.4f}, R²: {r2:.4f}")
```
